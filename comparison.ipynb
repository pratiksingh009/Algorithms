```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('output.csv')
data1 = data[:100]
```

```python
y1 = data1['actual']
y2 = data1['model1']
y3 = data1['model2']
```

```python
plt.figure(figsize=(8, 4))
plt.plot(y1, color='blue', label='Actual Price',marker='s')
plt.plot(y2, color='red', label='MLR Model 1',marker='^')
plt.plot(y3, color='green', label='MLR Model 2',marker='o')

# Add labels and title
plt.ylabel('Property Price')
# Add legend
plt.legend()
plt.savefig("comparison.png")
# Show plot
plt.show()
```
